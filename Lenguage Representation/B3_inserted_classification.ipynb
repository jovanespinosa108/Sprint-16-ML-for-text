Insertados para clasificacin

Ahora vamos a descubrir cmo la representacin vectorial puede ayudar a resolver tareas de clasificacin y regresin.

Digamos que hay un corpus de texto que necesita ser clasificado. En ese caso, nuestro modelo constar de dos bloques:

Modelos para convertir palabras en vectores: las palabras se convierten en vectores numricos.
Modelos de clasificacin: se utilizan vectores como caractersticas.




Vamos a repasar los detalles.

1. Antes de pasar a la vectorizacin de palabras, necesitaremos realizar un preprocesamiento:

- Cada texto est tokenizado (descompuesto en palabras).
- Luego se lematizan las palabras (reducidas a su forma raz). Sin embargo, los modelos ms complejos, como BERT, no requieren este paso porque entienden las formas de las palabras.
- El texto se limpia de palabras vacas o caracteres innecesarios.
- Para algunos algoritmos (por ejemplo, BERT), se agregan tokens especiales para marcar el comienzo y el final de las oraciones.

2. Cada texto adquiere su propia lista de tokens despus del preprocesamiento.
3. Luego, los tokens se pasan al modelo, que los vectoriza mediante el uso de un vocabulario de tokens precompilado. En la salida obtenemos vectores de longitud predeterminada formados para cada texto.
4. El paso final es pasar las caractersticas (vectores) al modelo. Luego, el modelo predice la tonalidad del texto: "0"  negativo o "1"  positivo.

 Insertados para clasificacin (versin sencilla)

Cuando queremos que una computadora clasifique texto (ejemplo: si una resea es positiva o negativa), no podemos darle directamente las palabras, porque las mquinas entienden nmeros, no letras.

Entonces seguimos este proceso:

Preprocesar el texto

Dividimos la oracin en palabras (tokenizacin).
Ejemplo: I love pizza  [I, love, pizza].

A veces reducimos las palabras a su raz (lematizacin): playing  play.

Quitamos palabras sin importancia (stopwords), como the, a, is.

En modelos como BERT, no hace falta lematizar y adems se agregan tokens especiales para marcar inicio y fin de la oracin.

Convertir palabras en vectores (embeddings)
Cada palabra (o el texto completo) se traduce a un vector numrico que captura su significado.
Ejemplo:

"love"  [0.23, -0.11, 0.87, ...]

"pizza"  [0.51, -0.44, 0.03, ...]

Clasificacin usando vectores
Esos vectores ahora son las caractersticas de entrada a un modelo de ML (como regresin logstica, red neuronal o BERT).
El modelo aprende a detectar patrones:

Vectores cercanos a good, love, great  positivo (1).

Vectores cercanos a bad, boring, terrible  negativo (0).

 Resumen en una frase

 Primero limpiamos y convertimos el texto en nmeros (embeddings), y luego usamos esos nmeros en un modelo que decide la clase (ej. positivo o negativo).
