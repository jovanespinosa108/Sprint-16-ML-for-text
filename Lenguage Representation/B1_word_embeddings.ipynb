### Insertados de palabras

Los desarrollos recientes en el análisis de texto toman en cuenta el contexto y el significado de las palabras al transformar el texto en vectores. Esto no es posible con el modelo "bolsa de palabras" o TF-IDF porque ambos modelos tratan cada palabra como un componente separado del texto.

Las máquinas no pueden trabajar directamente con palabras, imágenes y audio porque están diseñadas para trabajar con números. Los modelos de lenguaje transforman el texto en vectores y utilizan un conjunto de técnicas que se denominan colectivamente insertados de palabras. Esto significa que una palabra está embebida en un espacio vectorial que representa un modelo de lenguaje. Diferentes áreas en ese espacio vectorial tienen diferentes significados desde la perspectiva del lenguaje.

Un vector creado a partir de una palabra mediante el insertado de palabras llevará información contextual sobre la palabra y sus propiedades semánticas. De esta forma, no se pierde la definición convencional de una palabra y su significado en contexto.

Las propiedades semánticas son los componentes de una palabra que contribuyen a su significado. La palabra "marinero" tiene las siguientes propiedades semánticas: "masculino", "ocupación", "persona", "naval". Estas propiedades distinguen la palabra "marinero" de otras palabras. Pero las palabras, por lo general, no se sostienen solas y están rodeadas de otras palabras en una oración. Compara "El gato enredó todos los hilos" y "Perdí el hilo varias veces". La palabra "hilo" tiene diferentes significados según el contexto.

Este concepto permite trabajar con palabras en su forma vectorizada sin dejar de considerar el contexto. Esto significa que las palabras con contextos similares tendrán vectores similares.  La distancia (coseno, euclidiana, etc.) entre los vectores es una medida común de su similitud.

Para comprender los vectores similares y distantes, vamos a echar un vistazo a algunos vectores con una longitud igual a 2 para las palabras "pinzón", "frailecillo", "oso hormiguero" y "perezoso".

No importa cómo obtuvimos los vectores. Lo importante es comprender las propiedades que tienen estos vectores y cómo se relacionan con su contexto.

"Pinzón" y "frailecillo" están claramente más cerca el uno del otro que "oso hormiguero" y "perezoso". Tanto "pinzón" como "frailecillo" podrían encajar en esta oración: "El pico rojo del ___ brilló en el cielo azul". ¿Por qué? Porque tanto los pinzones como los frailecillos tienen pico, mientras que los osos hormigueros y los perezosos no.

Las palabras "oso hormiguero" y "perezoso" son más adecuadas en la oración: "Juan siempre contaba historias sobre animales extraños que encontró en Perú. Por ejemplo, un extraño _". Como puedes ver, la similitud de ambos pares de palabras se define por lo bien que estas pueden encajar en el contexto. Es más probable que los vectores cercanos encajen en las mismas oraciones.

Podemos suponer que la longitud del primer vector se asocia con la capacidad de volar y que la longitud del segundo vector se conecta con los mamíferos xenartros o la vellosidad. Sin embargo, no siempre podemos interpretar los vectores como características.

    Realmente no. Una representación vectorial con insertado de palabras mantiene el significado de las palabras con respecto al contexto, es decir, con enlaces contextuales con otras palabras. La bolsa de palabras no es capaz de producir tales vectores.

    Existen herramientas muy fáciles como la TF-IDF. Sin embargo, la calidad de los modelos creados con estas suele ser inferior a la de los creados con insertado de palabras.
