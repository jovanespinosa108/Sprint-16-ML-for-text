### Word2vec

Hablemos de word2vec (palabra a vector), un mtodo popular para construir representaciones de lenguaje.

Entonces, cmo funciona word2vec? Recordars que en la leccin anterior, las palabras "pinzn" y "frailecillo" se consideraron similares porque las dos se pueden usar en contexto con picos rojos, y las palabras "oso hormiguero" y "perezoso" se pueden usar en enunciados sobre Per. Es decir, el significado de las palabras depende del contexto.

Word2vec es un mtodo que se utiliza para convertir palabras en vectores (de ah su nombre) para que las palabras cercanas semnticamente acerquen los vectores entre s. Las libreras que ofrecen este mtodo suelen contener modelos word2vec entrenados previamente, con cada modelo entrenado en un corpus especfico de textos.

Por lo tanto, para obtener vectores relevantes para tus textos, debes seleccionar un modelo word2vec que se haya creado para un corpus de textos semnticamente relevante. Por ejemplo, para convertir noticias en vectores, se debe usar un modelo word2vec entrenado en un corpus de noticias. Puedes consultar la lista de modelos capaces de convertir palabras en ingls en vectores enviados con spaCy (materiales en ingls) en el enlace https://spacy.io/models/en (materiales en ingls) (comprueba los que tengan definida la propiedad "Vectors"). La otra librera popular para implementar el mtodo word2vec es gensim (materiales en ingls), y la puedes consultar en el enlace (materiales en ingls). Esto es solo para darte una idea general de cmo se ven los modelos previamente entrenados.

Adems de convertir palabras en vectores, los modelos word2vec tambin se pueden usar para resolver tareas especficas de PNL. Por ejemplo, pueden ayudar a predecir si algunas palabras son vecinas o no. Las palabras se consideran vecinas si caen en la misma "ventana" (distancia mxima entre palabras). Dado que cada palabra en un par de palabras es una caracterstica, esto significa que hay dos caractersticas en cada par. El objetivo es averiguar si las palabras del par son vecinas o no.

Vamos a usar el enunciado sobre el "puffin" (frailecillo) como el conjunto de datos para nuestra tarea:

- Red beak of the puffin flashed in the blue sky

Aqu est el mismo texto despus de la lematizacin:

- red beak puffin flash in blue sky

La palabra que predecimos en el enunciado es "puffin". Las palabras vecinas son las ms cercanas de ambos lados: "red", "beak", "flash" y "in". Por ejemplo, estas cinco palabras forman un cincograma:



Ahora tenemos cuatro pares de vecinas:

- puffin    red
- puffin    beak
- puffin    flash
- puffin    in

Si tuviramos que predecir la palabra "flash", las palabras vecinas seran: "beak," "puffin," "in," and "blue".



Ahora tenemos cuatro nuevos pares de palabras vecinas:

- puffin    red
- puffin    beak
- puffin    flash
- puffin    in
- flash    beak
- flash    puffin
- flash    in
- flash    blue

Podemos "desplazar la ventana" a travs de todo el texto y extraer la lista completa de palabras vecinas.

Otra cosa que puede hacer word2vec es entrenar un modelo para distinguir pares de vecinas verdaderas de las aleatorias. Esta tarea es como una tarea de clasificacin binaria, donde las caractersticas son palabras y el objetivo es la respuesta a la pregunta de si las palabras son vecinas verdaderas o no.

Ya tenemos nuestros ejemplos positivos, pero para obtener los negativos necesitaremos tomar palabras aleatorias del corpus y emparejarlas:

- puffin juan
- beak bizarre
- sky animal
- red tell

Ahora tenemos un conjunto de entrenamiento. Aqu, "1" indica que las palabras son verdaderas vecinas, mientras que "0" significa que no son vecinas.

Pregunta

Por qu necesitamos ejemplos negativos de pares de palabras?

* Para que el modelo sea capaz de diferenciar los ejemplos positivos de los negativos.
    Correcto. Los ejemplos positivos por s solos no son suficientes para entrenar un modelo.

Este es un mtodo de regularizacin. Debido a los pares aleatorios, el modelo se vuelve ms estable.

Este es un tipo de bootstrapping. Las submuestras aleatorias ayudan a evitar el sobreajuste.
